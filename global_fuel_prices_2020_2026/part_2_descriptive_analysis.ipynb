Top 5 pays les plus chers pour l’essence en 2025.

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df_2025 = df[df["date"].dt.year ==2025]
df_2025[["date","country","petrol_usd_liter"]]
```

```python
df_2025.groupby("country")["petrol_usd_liter"].mean().sort_values().head(5)
```

```python
df_2025.groupby("country")["petrol_usd_liter"].mean().sort_values(ascending=False).head(5)
```
