Impact des subventions : comparer les pays avec subventions vs sans subventions.

Taxes : analyser si les pays avec forte taxation ont systématiquement des prix plus élevés.

```python
df["tax_group"] = pd.cut(
    df["tax_percentage"],
    bins=[0,15,30,100],
    labels = ["Low","Medium","Hight"],
    include_lowest=True)
    
```

```python
df.groupby("tax_group")["petrol_usd_liter"].mean()
```

```python
df.groupby("tax_group")["diesel_usd_liter"].mean()
```

```python
df.groupby("tax_group")["lpg_usd_liter"].mean()
```
