Classement dynamique : chaque année, identifier le pays le plus cher et le moins cher.

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df["year"]= df["date"].dt.year
```

```python
df.groupby(["year","country"])["petrol_usd_liter"].mean().reset_index().loc[
    lambda x: x.groupby("year")["petrol_usd_liter"].idxmax()]
```

```python
df.groupby(["year","country"])["petrol_usd_liter"].mean().reset_index().loc[
    lambda x: x.groupby("year")["petrol_usd_liter"].idxmin()]
```
